```python
# Step 1: Import required libraries

import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Downloading Stock Market Data

# Define the ticker symbols for the three banks
tickers = ["HDFCBANK.NS", "ICICIBANK.NS", "KOTAKBANK.NS"]

# Fetch historical data for the year 2024
data = yf.download(tickers, start="2024-01-01", end="2024-12-30", group_by='ticker')

# Display the first 5 rows of the dataset
print("First 5 rows of the dataset:")
print(data.head())

# Step 3: Basic Data Exploration

# Check the shape of the dataset
print("\nShape of the dataset:")
print(data.shape)

# Check column names
print("\nColumn names:")
print(data.columns)

# Summary statistics for HDFC Bank
hdfc_data = data['HDFCBANK.NS']
print("\nSummary statistics for HDFC Bank:")
print(hdfc_data.describe())

# Calculate daily returns for HDFC Bank
hdfc_data['Daily Return'] = hdfc_data['Close'].pct_change()

# Summary statistics for ICICI Bank
icici_data = data['ICICIBANK.NS']
print("\nSummary statistics for ICICI Bank:")
print(icici_data.describe())

# Calculate daily returns for ICICI Bank
icici_data['Daily Return'] = icici_data['Close'].pct_change()

# Summary statistics for Kotak Mahindra Bank
kotak_data = data['KOTAKBANK.NS']
print("\nSummary statistics for Kotak Mahindra Bank:")
print(kotak_data.describe())

# Calculate daily returns for Kotak Mahindra Bank
kotak_data['Daily Return'] = kotak_data['Close'].pct_change()

# Step 4: Plot the closing price and daily returns for all three banks

plt.figure(figsize=(12, 12))

# Plot for HDFC Bank
plt.subplot(3, 2, 1)
hdfc_data['Close'].plot(title="HDFC Bank - Closing Price")
plt.subplot(3, 2, 2)
hdfc_data['Daily Return'].plot(title="HDFC Bank - Daily Returns", color='orange')

# Plot for ICICI Bank
plt.subplot(3, 2, 3)
icici_data['Close'].plot(title="ICICI Bank - Closing Price")
plt.subplot(3, 2, 4)
icici_data['Daily Return'].plot(title="ICICI Bank - Daily Returns", color='orange')

# Plot for Kotak Mahindra Bank
plt.subplot(3, 2, 5)
kotak_data['Close'].plot(title="Kotak Mahindra Bank - Closing Price")
plt.subplot(3, 2, 6)
kotak_data['Daily Return'].plot(title="Kotak Mahindra Bank - Daily Returns", color='orange')

plt.tight_layout()
plt.show()

# Step 5: Saving the Processed Data to CSV Files

# Save the data for each bank to separate CSV files
hdfc_data.to_csv('hdfc_bank_data.csv')
icici_data.to_csv('icici_bank_data.csv')
kotak_data.to_csv('kotak_bank_data.csv')

print("\nStock data saved to CSV files.")
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Define tickers and download data
tickers = ["HDFCBANK.NS", "ICICIBANK.NS", "KOTAKBANK.NS"]
data = yf.download(tickers, start="2024-01-01", end="2024-12-30", group_by='ticker')

# Calculate daily returns and print summary statistics
for ticker in tickers:
    stock_data = data[ticker]
    stock_data['Daily Return'] = stock_data['Close'].pct_change()
    print(f"\nSummary statistics for {ticker}:")
    print(stock_data.describe())

# Plot the closing price and daily returns
plt.figure(figsize=(12, 12))
for i, ticker in enumerate(tickers, 1):
    stock_data = data[ticker]
    plt.subplot(3, 2, 2*i-1)
    stock_data['Close'].plot(title=f"{ticker} - Closing Price")
    plt.subplot(3, 2, 2*i)
    stock_data['Daily Return'].plot(title=f"{ticker} - Daily Returns", color='orange')

plt.tight_layout()
plt.show()

# Save to CSV
for ticker in tickers:
    data[ticker].to_csv(f'{ticker.lower()}_data.csv')

print("\nStock data saved to CSV files.")
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Define ticker symbols
tickers = ["HDFCBANK.NS", "ICICIBANK.NS", "KOTAKBANK.NS"]

# Fetch historical data
data = yf.download(tickers , start="2024-01-01", end="2024-12-30", group_by='ticker')

# Calculate daily returns safely using .loc
for ticker in tickers:
    data.loc[:, (ticker, 'Daily Return')] = data[ticker]['Close'].pct_change()

# Plot closing prices
plt.figure(figsize=(12, 6))
for i, ticker in enumerate(tickers):
    plt.subplot(3, 1, i+1)
    data[ticker]['Close'].plot(title=f"{ticker} - Closing Price")

plt.tight_layout()
plt.show()

# Plot daily returns
plt.figure(figsize=(12, 6))
for i, ticker in enumerate(tickers):
    plt.subplot(3, 1, i+1)
    data[ticker]['Daily Return'].plot(title=f"{ticker} - Daily Returns", color='orange')

plt.tight_layout()
plt.show()
```
